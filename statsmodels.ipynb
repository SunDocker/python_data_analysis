```python
import numpy as np
import statsmodels.api as sm
# 1 生成四维数据点
# 1.1 生成自变量
x1 = np.random.normal(0, 1, 1000)
x2 = np.random.normal(0, 1.5, 1000)
x3 = np.random.normal(0, 2, 1000)
X = np.c_[x1, x2, x3]
# 1.2 生成因变量
eps = np.random.normal(0, 0.3, 1000) # 噪声数据
c = [1, 2, 3]
y = X @ c + eps
# 2 利用生成的数据点拟合
model = sm.OLS(y, sm.add_constant(X))
res = model.fit()
res.summary()
```
